```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import stft
```

```python
FPS = 1000
t = np.arange(0, 1, 1/FPS)
x1 = np.sin(2*np.pi*100*t)
x2 = np.sin(2*np.pi*200*t)
x3 = np.sin(2*np.pi*300*t)
x4 = np.sin(2*np.pi*400*t)
xtot = np.concatenate([x1, x2, x3, x4])
```

```python
frez, trez, zrez = stft(xtot, fs=FPS, window='hann', nperseg=100)

plt.figure()
plt.pcolormesh(trez, frez, np.abs(zrez), shading='gouraud')
plt.show()
```
