```python
import numpy as np
import matplotlib.pyplot as plt

sh = (20, 10)
M = np.ones(sh)
M /= 20 #np.linalg.norm(M)
rank = []
cond = []
eigs = []
for i in range(5000):
    rank += [np.linalg.matrix_rank(M)]
    cond += [np.linalg.cond(M)]
    eigs += [np.sort(np.real(np.linalg.eig(M.T.dot(M))[0]))]
    dM = 0.01 * np.random.normal(0, 1, sh)
    M += dM
    M[M < 0] = 0
    
    
plt.figure()
plt.semilogy(rank, label='rank')
plt.semilogy(cond, label='cond')
plt.legend()
plt.ylim([1.0e-2, 1.0e+3])
plt.show()
plt.figure()
plt.semilogy(eigs)
plt.ylim([1.0e-5, 1.0e+2])
plt.show()

print(eigs[-1])
```
