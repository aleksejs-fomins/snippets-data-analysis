```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gainz(p, n):
    return (p**n - 1) / (p - 1) / n
    
nArr = np.arange(1, 40)
    
plt.figure(figsize=(5,5))
for p in np.linspace(1.01, 1.08, 8):
    
    rez = [gainz(p, n) for n in nArr]
    plt.plot(nArr, rez, label=str(p))
    print(p, rez[-1])
    
plt.legend()
plt.title('Relative gain from investing same amount each year for fixed percent')
plt.ylabel('Relative gain')
plt.xlabel('Years')
plt.show()
```
