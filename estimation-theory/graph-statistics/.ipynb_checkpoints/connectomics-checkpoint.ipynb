```python
import matplotlib.pyplot as plt
import numpy as np
import networkx as nx
```

```python
pLst = np.linspace(0.01, 0.9, 20)
diam = np.zeros(len(pLst))

for iP in range(len(pLst)):
    for iAvg in range(100):
        M = (np.random.uniform(0, 1, 100) <= pLst[iP]).reshape((10, 10))
        edges = [(i, j) for i in range(10) for j in range(10) if M[i][j]]
    #     plt.figure()
    #     plt.imshow(aaa)

        G = nx.Graph()
        G.add_edges_from(edges)
    #     print(len(edges), G.number_of_edges())
        try:
            diam[iP] += nx.diameter(G) / 100
        except:
            diam[iP] += 11 / 100
    
plt.figure()
plt.plot(pLst, diam);
```

```python
MATSIZE = 100

punishMatrix = [[(i - j)**2 for j in range(MATSIZE)] for i in range(MATSIZE)]
# plt.figure()
# plt.imshow(punishMatrix);

def diagonality(M):
    return np.sum(np.multiply(M, punishMatrix))

def swapRows(M, i, j):
    tmp = np.copy(M[i])
    M[i] = M[j]
    M[j] = tmp
    
def swapCols(M, i, j):
    tmp = np.copy(M[:,i])
    M[:,i] = M[:,j]
    M[:,j] = tmp

# MCONN = (np.random.uniform(0, 1, MATSIZE**2) <= 0.2).reshape((MATSIZE, MATSIZE))
# MVAL = np.random.uniform(0, 1, MATSIZE**2).reshape((MATSIZE, MATSIZE))
# M = np.multiply(MCONN, MVAL)
M = np.random.uniform(0, 1, MATSIZE**2).reshape((MATSIZE, MATSIZE))
plt.figure()
plt.imshow(M);

diagThis = diagonality(M)
print(diagThis)

diagHist = [diagThis]

SIMTIME = 100000
swapSwitchLst = [int(i / 1000) % 2 for i in range(SIMTIME)]

for i in range(SIMTIME):
    swapSwitch = swapSwitchLst[i]#np.random.randint(0, 2) == 0    
    swapIdx = np.random.randint(0, MATSIZE, 2)
    swapRows(M, swapIdx[0], swapIdx[1]) if swapSwitch else swapCols(M, swapIdx[0], swapIdx[1])
    diagNew = diagonality(M)
    if (diagNew < diagThis):# or (np.random.uniform(0, 1) > 0.99):
        diagThis = diagNew
    else:
        swapRows(M, swapIdx[0], swapIdx[1]) if swapSwitch else swapCols(M, swapIdx[0], swapIdx[1])
        
    diagHist += [diagThis]

            
plt.figure()
plt.imshow(M);
print(diagThis)

plt.figure()
plt.loglog(diagHist);
```
