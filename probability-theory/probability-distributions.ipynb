```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta, binom, gamma, cauchy

import ipywidgets as widgets
from ipywidgets import interact, interact_manual
```

```python
a, b = 2.31, 0.627
mean, var, skew, kurt = beta.stats(a, b, moments='mvsk')
np.array([mean, var, skew, kurt])
```

```python
%matplotlib notebook
fig, ax = plt.subplots(figsize=(2,2))

x = np.linspace(0, 1, 100)
line, = ax.plot(x, beta.pdf(x, 1, 1), 'r-')
plt.show()

@interact
def plotdist(mu=(0, 1, 0.01), k=(1, 10,0.01)):
    print(a, b)
    line.set_ydata(beta.pdf(x, mu*k, (1-mu)*k))
    ax.relim()
    ax.autoscale_view()
    fig.canvas.draw()
    fig.canvas.flush_events()
```
