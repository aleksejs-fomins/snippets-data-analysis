# Metrics - Granger Causality

http://www.statsmodels.org/stable/generated/statsmodels.tsa.stattools.grangercausalitytests.html

```python
import numpy as np
import statsmodels.tsa.stattools as stattools
import matplotlib.pyplot as plt

x = np.linspace(0, 2, 200)
# y = np.sin(2*np.pi*x)
# z = np.cos(2*np.pi*x)
y = np.random.uniform(-1, 1, 200)
z = np.random.uniform(-1, 1, 200)

yz = np.vstack((y,z)).T
rezGranger = stattools.grangercausalitytests(yz, 50, verbose=False)

print(rezGranger)
```

```python
rezfit = [rezGranger[i][0]['ssr_chi2test'][0] for i in range(1,50)]

plt.figure()
plt.plot(rezfit)
plt.show()
```
