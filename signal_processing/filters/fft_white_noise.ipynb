```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
aaa = np.random.exponential(1, 1000)

plt.figure()
plt.plot(aaa, '.')
plt.show()
```

```python
fft = np.fft.fft(aaa)

fig, ax = plt.subplots(nrows=2)
ax[0].plot(np.real(fft)[1:])
ax[1].plot(np.imag(fft)[1:])
plt.show()
```

```python
plt.figure()
plt.plot(np.real(fft)[1:] - np.flip(np.real(fft)[1:]))
plt.show()
```

```python
npoint = (10**np.linspace(1,4,20)).astype(int)
fftLst = []

for n in npoint:
    aaa = np.random.exponential(1, n)
    fft = np.fft.fft(aaa)
    fftLst += [np.real(fft)[0]]
    
plt.figure()
plt.semilogx(npoint, np.array(fftLst) / npoint)
plt.show()
```
