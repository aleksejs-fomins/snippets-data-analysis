Observations:
* Random matrix of zero mean produces eigenvalue spectrum which is a ball around origin in RE-IM space
* Radius of eigenvalues of the random matrix increases with matrix size and with variance
* Having nonzero mean produces a few extra real eigenvalues outside the ball

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
eigL = np.zeros(0)
for i in range(20):
    M = np.random.normal(0.5, 3, size=(200,200))
#     plt.figure()
#     plt.imshow(M)

    evals, evecs = np.linalg.eig(M.T)
    # idxsort = np.argsort(evals)
    
    eigL = np.hstack((eigL, evals))

plt.figure()
plt.plot(np.real(eigL), np.imag(eigL), 'o');
```
