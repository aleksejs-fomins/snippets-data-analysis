## Synapses

```python
# Regular spiking cells (will be useful as stimulation for other neurons later)
start_scope()

# Fake spike generator
input_spikes = SpikeGeneratorGroup(10, np.arange(10), np.arange(10)*ms, period=10*ms)

# Generate some spikes and check that it does what we think it should
spike_mon  = SpikeMonitor(input_spikes)
run(100*ms)
brian_plot(spike_mon);
```

```python
start_scope()

# Reuse the previous model

Cm = 200*pF    # Membrane capacitance
E_L = -70*mV   # Potential
E_Th = -45*mV  # Spike threshold
f = 100*Hz     # Input current frequency

# NOTE_1: we define Vm using below diff.eq. The associated units (volt) are the units of Vm by construction
# NOTE_2: variables that are unique for each neuron must have separate equations (like g_L)
eqs = '''
dVm/dt = (I_inj + g_L*(E_L - Vm)) / Cm : volt (unless refractory)
g_L : siemens
I_inj = 0*nA : amp
'''

group = NeuronGroup(10, eqs, 
                    threshold='Vm>E_Th',  # When to spike
                    reset='Vm=E_L',       # What to do when spike occurs
                    refractory=5*ms,      # Refractory period
                    method='euler')       #, method=exact)
group.Vm = E_L
group.g_L = '10*nS + rand()*5*nS'  # Define random leak conductance for each neuron (nS - nano Siemens)

# Generate some synapses between our neurons
# Most trivial gap junction-like synapses, that change post-synaptic voltage
input_spikes = SpikeGeneratorGroup(10, np.arange(10), np.arange(10)*ms, period=10*ms)
syn = Synapses(input_spikes, group, 'w : volt',
               on_pre='Vm_post += w')

syn.connect()
syn.w = '1*mV + rand()*4*mV'#1*mV
syn.delay = '1*ms + rand()*5*ms'

mon = StateMonitor(group, 'Vm', record=True)
# spike_mon = SpikeMonitor(group)

run(100*ms, report='text')
```
