## Brian's unit system

All the standard SI units are available using their name (volt, siemens, meter (or metre), second, etc.), together with standard prefixes n (nano), u (micro), m (milli), k (kilo), M (mega), etc.

Because it would lead to clashes with user-defined variable names too often, single letter names are *not* defined (e.g. no `s` or `m`), but for convenience the most commonly used abbreviations for neuroscience are available (e.g. `ms` instead of `msecond`, `mV` instead of `mvolt`, `nS` instead of `nsiemens`, etc.)

## Neural dynamics

Membrane potential equation:
$$
C_m\frac{d V_m}{dt} = I + g_L (E_L - V_m)
$$

````python
start_scope()

Cm = 200*pF    # Membrane capacitance
E_L = -70*mV   # Potential
E_Th = -45*mV  # Spike threshold
I0 = 5.0*nA    # Input current magnutude
f = 100*Hz     # Input current frequency

# NOTE_1: we define Vm using below diff.eq. The associated units (volt) are the units of Vm by construction
# NOTE_2: variables that are unique for each neuron must have separate equations (like g_L)
eqs = '''
dVm/dt = (I_inj + g_L*(E_L - Vm)) / Cm : volt (unless refractory)
g_L : siemens
I_inj = I0*sin(2 * pi * f * t) : amp
'''

group = NeuronGroup(10, eqs, 
                    threshold='Vm>E_Th',  # When to spike
                    reset='Vm=E_L',       # What to do when spike occurs
                    refractory=5*ms,      # Refractory period
                    method='euler')       #, method=exact)
group.Vm = E_L
group.g_L = '10*nS + rand()*5*nS'  # Define random leak conductance for each neuron (nS - nano Siemens)

# mon = StateMonitor(group, 'Vm', record=True)
mon = StateMonitor(group, 'Vm', record=[0, 5, 7])
spike_mon = SpikeMonitor(group)

run(100*ms, report='text')
````

````python
# plot membrane potential over time
plt.figure()
plot1 = plot(mon.t/ms, mon.Vm.T / mV);
plt.figure()
plot(spike_mon.t/ms, spike_mon.i, '|');
````
