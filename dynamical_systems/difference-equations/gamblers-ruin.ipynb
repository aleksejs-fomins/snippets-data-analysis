## Gambler's ruin - ODE approximation

### Winning probabilities

Standard Gambler's ruin is given by the difference equation for the probability of win

$w(x) = p w(x-\Delta x) + q w(x+\Delta x)$

where $p \in [0, 1]$, $q = 1-p$. Denote $w_0 = w(t)$, $w_+ = w(x+\Delta x)$, $w_- = w(x - \Delta x)$.
After some algebraic manipulation

$w_{+} + w_{-} -2 w_0 = -(1 - 2p) (w_{+} - w_{-})$

which can be approximated by an ODE

$\ddot{w} \approx -2(1-2p)\Delta x \dot{w} = -\zeta \dot{w}$

Note that we must explicitly force $zeta$ to be finite, otherwise the function stops being continuous. The solution is

$w(x) = A + Be^{-\zeta x}$

Plugging in the boundary conditions $w(0) = 0$, $w(1) = 1$, we get

$w(x) = \frac{1 - e^{-\zeta x}}{1 - e^{-\zeta}}$

For a balanced game ($\zeta \rightarrow 0$) the expression converges to $w(x) = x$

### Average time until win or bust

Difference equation for the time of win/bust is given by

$T_0 = p T_{-} + q T_{+} + 1$

After some algebraic manipulation, we obtain an ODE

$\ddot{T} \approx -2(1-2p)\Delta x \dot{T} -\frac{2}{\Delta x^2} = -\zeta \dot{T} - \eta$

Note that we must also force the gain $\eta$ to be finite for the distribution of $T$ to be continuous. We already know the homogeneous solution for the equation from the above proof. By guesswork, the particular solution to this equation can be found to be

$T_p(x) = -\frac{\eta}{\zeta}x + K$

for some constant $K$. Thus, the full solution is

$T(x) = T_h(x) + T_{p} =  A + Be^{-\zeta x} -\frac{\eta}{\zeta}x$

Plugging in the boundary conditions $T(0) = 0$, $T(1) = 0$, we get

$T(x) = \frac{\eta}{\zeta}\bigl ( \frac{1 - e^{-\zeta x}}{1 - e^{-\zeta}} - x \bigr )$

For a balanced game ($\zeta \rightarrow 0$) the expression converges to

$T(x) = -\zeta \frac{t(t-1)}{2}$

```python
import numpy as np
import matplotlib.pyplot as plt

def T(x, z):
    return ((1 - np.exp(-z*x))/((1 - np.exp(-z))) - x)/z
x = np.linspace(0, 1, 100)
plt.figure()
plt.plot(x, T(x, 1))
plt.plot(x, T(x, -1))
plt.plot(x, T(x, 5))
plt.plot(x, T(x, -5))
plt.show()
```
