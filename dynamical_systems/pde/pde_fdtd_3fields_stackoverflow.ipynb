Original equations
$$
\begin{eqnarray}
    \frac{\partial H}{\partial x}+\beta_h (H - W) &=& 0 \\
    \frac{\partial C}{\partial y} + \beta_c (C - W) &=& 0 \\
    \lambda_h \frac{\partial^2 W}{\partial x^2} + \lambda_c V\frac{\partial^2 W}{\partial y^2}
    - \frac{\partial H}{\partial x} - V\frac{\partial C}{\partial y} &=& 0
\end{eqnarray}
$$

Central differences
$$
\begin{eqnarray}
    H_{i+1, j} - H_{i-1, j} + 2\Delta x\beta_h H_{ij} &=& 2\Delta x\beta_h W_{ij} \\
    C_{i, j+1} - C_{i, j-1} + 2\Delta y\beta_c C_{ij} &=& 2\Delta y\beta_c W_{ij} \\
      \lambda_h (W_{i+1, j} - 2W_{i,j} + W_{i-1, j})
    + \lambda_c V \eta^2 (W_{i, j+1} - 2W_{i,j} + W_{i, j-1})
    + 2(\beta_h + V \beta_c)\Delta x^2 W_{ij}
    &=& 2\beta_h \Delta x^2 H_{ij} + 2V\beta_c \Delta x^2 C_{ij}
\end{eqnarray}
$$

Constraints
$$\theta_w(0,y)=1, \theta_w(x,0)=0$$
$$\frac{\partial \theta_w(1,y)}{\partial x}=\frac{\partial \theta_w(x,1)}{\partial y}=0$$
$$\theta_h(0,y)=1, \theta_c(x,0)=0$$

* (@i = 1) : $W_{i-1,j} = 1$, $H_{i-1,j} = 1$
* (@j = 1) : $W_{i,j-1} = 0$, $H_{i,j-1} = 0$
* (@i = n) : $W_{i-1,j} = W_{i+1,j}$
* (@j = n) : $W_{i,j-1} = W_{i,j+1}$

Problem: Need to find 2nd boundary condition for $H$ and $C$. Use trick - take second derivative
$$
\begin{eqnarray}
    \partial_x^2 H + \beta_h (\partial_x H - \partial_x W) &=& 0 \\
    \partial_y^2 C + \beta_c (\partial_y C - \partial_y W) &=& 0
\end{eqnarray}
$$
Discretize
$$
\begin{eqnarray}
    (H_{i+1,j} - 2H_{i,j} + H_{i-1,j}) + \frac{\beta_h \Delta X}{2} (H_{i+1, j} - H_{i-1, j})
        &=& \frac{\beta_h \Delta X}{2} (W_{i+1, j} - W_{i-1, j}) \\
    (C_{i,j+1} - 2C_{i,j} + C_{i,j-1}) + \frac{\beta_c \Delta Y}{2} (C_{i, j+1} - C_{i, j-1})
        &=& \frac{\beta_h \Delta Y}{2} (W_{i, j+1} - W_{i, j-1}) \\
\end{eqnarray}
$$

At the outer edge derivatives of W are zero. Can solve for the outer edge of $H$ and $C$

$$
\begin{eqnarray}
    (\frac{\beta_h \Delta X}{2} + 1) H_{i+1, j} &=& 2H_{i,j} - (1 - \frac{\beta_h \Delta X}{2}) H_{i-1,j} \\
    (\frac{\beta_c \Delta Y}{2} + 1) C_{i, j+1} &=& 2C_{i,j} - (1 - \frac{\beta_c \Delta Y}{2}) C_{i,j-1}
\end{eqnarray}
$$
Thus, at the outer boundaries, we have
$$
\begin{eqnarray}
    (\frac{\beta_h^2 \Delta X^2}{2} + \beta_h \Delta c + 1) H_{i,j} - H_{i-1,j}
        &=& \Delta x\beta_h (\frac{\beta_h \Delta c}{2} + 1) W_{ij} \\
    (\frac{\beta_c^2 \Delta Y^2}{2} + \beta_c \Delta y + 1) C_{i,j} - C_{i,j-1}
        &=& \Delta y\beta_c (\frac{\beta_c \Delta y}{2} + 1) W_{ij} \\
\end{eqnarray}
$$
