```python
import numpy as np
import matplotlib.pyplot as plt

N_DISCR = 1000
rand = np.random.normal(0,1, N_DISCR)
markov = np.zeros(N_DISCR+1)
for i in range(N_DISCR):
    markov[i+1] = markov[i]+rand[i]

alpha = 0.1
runMean = np.zeros(N_DISCR+1)
runVar = np.zeros(N_DISCR+1)

for i in range(N_DISCR):
    runMean[i+1] = (1-alpha)*runMean[i] + alpha*markov[i]
    runVar[i+1] = (1-alpha)*runVar[i] + alpha*(runMean[i] - markov[i])**2
       
plt.figure(figsize=(15,10))
plt.plot(markov)
plt.plot(runMean)
plt.plot(runMean + np.sqrt(runVar))
plt.plot(runMean - np.sqrt(runVar))
plt.show()
```
