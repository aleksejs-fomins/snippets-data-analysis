# I.I.D.

How can one check if the data is i.i.d? Is it possible? How efficient it is?

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats
```

### Test 1: Check asymptotic behaviour of mean and variance - both should converge

```python
y = np.random.normal(0, 1, 10000)
mean = [np.average(y[:i]) for i in range(1, 10001)]
std = [np.std(y[:i]) for i in range(1, 10001)]

plt.figure()
plt.plot(mean)
plt.plot(std)
plt.show()

y = np.random.normal(0, 1, 10000)
for i in range(1,10000):
    y[i] += y[i-1]

mean = [np.average(y[:i]) for i in range(1, 10001)]
std = [np.std(y[:i]) for i in range(1, 10001)]

plt.figure()
plt.plot(mean)
plt.plot(std)
plt.show()
```
